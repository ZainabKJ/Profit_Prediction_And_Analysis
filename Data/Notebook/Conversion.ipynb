```python
import pandas as pd
import os
```

```python
# Excel File (Benz)
excel_bal_b = '../ExcelFiles/Benz-Balance.xlsx'
excel_cf_b = '../ExcelFiles/Benz-CF.xlsx'
excel_is_b = '../ExcelFiles/Benz-IS.xlsx'
excel_r_b = '../ExcelFiles/Benz-R.xlsx'

# Excel File (Ford)
excel_bal_f = '../ExcelFiles/FORD-Balance.xlsx'
excel_cf_f = '../ExcelFiles/FORD-CF.xlsx'
excel_is_f = '../ExcelFiles/FORD-IS.xlsx'
excel_r_f = '../ExcelFiles/FORD-R.xlsx'

# Excel File (Tesla)
excel_bal_t = '../ExcelFiles/Tesla-Balance.xlsx'
excel_cf_t = '../ExcelFiles/Tesla-CF.xlsx'
excel_is_t = '../ExcelFiles/Tesla-IS.xlsx'
excel_r_t = '../ExcelFiles/Tesla-R.xlsx'

# CSV File (Benz)
output_B = '../csv2/Benz/'

# CSV File (Ford)
output_F = '../csv2/Ford/'

# CSV File (Tesla)
output_T = '../csv2/Tesla/'

# Creation of new folder for each file
os.makedirs(output_B, exist_ok=True)
os.makedirs(output_F, exist_ok=True)
os.makedirs(output_T, exist_ok=True)

# Read Excel files for all comapnies
Frame_b1 = pd.read_excel(excel_bal_b)
Frame_b2 = pd.read_excel(excel_cf_b)
Frame_b3 = pd.read_excel(excel_is_b)
Frame_b4 = pd.read_excel(excel_r_b)

Frame_f1 = pd.read_excel(excel_bal_f)
Frame_f2 = pd.read_excel(excel_cf_f)
Frame_f3 = pd.read_excel(excel_is_f)
Frame_f4 = pd.read_excel(excel_r_f)

Frame_t1 = pd.read_excel(excel_bal_t)
Frame_t2 = pd.read_excel(excel_cf_t)
Frame_t3 = pd.read_excel(excel_is_t)
Frame_t4 = pd.read_excel(excel_r_t)

# Creation of path for the CSV file (Benz)
csv_B_bal = os.path.join(output_B, 'Benz-Balance.csv')
csv_B_cf = os.path.join(output_B, 'Benz-CF.csv')
csv_B_is = os.path.join(output_B, 'Benz-IS.csv')
csv_B_r = os.path.join(output_B, 'Benz-R.csv')

# Creation of path for the CSV file (Ford)
csv_F_bal = os.path.join(output_F, 'Ford-Balance.csv')
csv_F_cf = os.path.join(output_F, 'Ford-CF.csv')
csv_F_is = os.path.join(output_F, 'Ford-IS.csv')
csv_F_r = os.path.join(output_F, 'Ford-R.csv')

# Creation of path for the CSV file(Tesla)
csv_T_bal = os.path.join(output_T, 'Tesla-Balance.csv')
csv_T_cf = os.path.join(output_T, 'Tesla-CF.csv')
csv_T_is = os.path.join(output_T, 'Tesla-IS.csv')
csv_T_r = os.path.join(output_T, 'Tesla-R.csv')

# Write DataFrame to CSV for all
Frame_b1.to_csv(csv_B_bal, index=False)
Frame_b2.to_csv(csv_B_cf, index=False)
Frame_b3.to_csv(csv_B_is, index=False)
Frame_b4.to_csv(csv_B_r, index=False)

Frame_f1.to_csv(csv_F_bal, index=False)
Frame_f2.to_csv(csv_F_cf, index=False)
Frame_f3.to_csv(csv_F_is, index=False)
Frame_f4.to_csv(csv_F_r, index=False)

Frame_t1.to_csv(csv_T_bal, index=False)
Frame_t2.to_csv(csv_T_cf, index=False)
Frame_t3.to_csv(csv_T_is, index=False)
Frame_t4.to_csv(csv_T_r, index=False)

# Print confirmation of Completetion
print(f'Conversion completed. CSV file saved at: {csv_B_bal}')
print(f'Conversion completed. CSV file saved at: {csv_B_cf}')
print(f'Conversion completed. CSV file saved at: {csv_B_is}')
print(f'Conversion completed. CSV file saved at: {csv_B_r}')

print(f'Conversion completed. CSV file saved at: {csv_F_bal}')
print(f'Conversion completed. CSV file saved at: {csv_F_cf}')
print(f'Conversion completed. CSV file saved at: {csv_F_is}')
print(f'Conversion completed. CSV file saved at: {csv_F_r}')

print(f'Conversion completed. CSV file saved at: {csv_T_bal}')
print(f'Conversion completed. CSV file saved at: {csv_T_cf}')
print(f'Conversion completed. CSV file saved at: {csv_T_is}')
print(f'Conversion completed. CSV file saved at: {csv_T_r}')
```
