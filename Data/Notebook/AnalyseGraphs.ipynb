```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load all data for all 3 companies
fordMergeData = pd.read_csv('../csv2/fMerge.csv') 
teslaMergeData = pd.read_csv('../csv2/tMerge.csv') 
benzMergeData = pd.read_csv('../csv2/mMerge.csv') 
```

```python
# Replace and empty values with 0
fordMergeData.fillna(0, inplace=True)
teslaMergeData.fillna(0, inplace=True)
benzMergeData.fillna(0, inplace=True)
```

```python
# In all 3 copmpanies convert the date values to date-time format
fordMergeData['Date'] = pd.to_datetime(fordMergeData['Quaterly Data'])
benzMergeData['Date'] = pd.to_datetime(benzMergeData['Quaterly Data'])
teslaMergeData['Date'] = pd.to_datetime(teslaMergeData['Quaterly Data'])
```

```python
# Plotting the Net Profit Margin, alongside the dates, on the graphs for Comapny 1 (Ford)
plt.figure(figsize=(10, 6))
plt.plot(fordMergeData['Date'], fordMergeData['Net Profit Margin'], marker='o', linestyle='-')
plt.title('Net Profit Margin - (F)')
plt.xlabel('Date')
plt.ylabel('Net Profit Margin')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plotting the Net Profit Margin, alongside the dates, on the graphs for Comapny 2 (Mercedes-Benz)
plt.figure(figsize=(10, 6))
plt.plot(benzMergeData['Date'], benzMergeData['Net Profit Margin'], marker='o', linestyle='-')
plt.title('Net Profit Margin - (M)')
plt.xlabel('Date')
plt.ylabel('Net Profit Margin')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Plotting the Net Profit Margin, alongside the dates, on the graphs for Comapny 3 (Tesla)
plt.figure(figsize=(10, 6))
plt.plot(teslaMergeData['Date'], teslaMergeData['Net Profit Margin'], marker='o', linestyle='-')
plt.title('Net Profit Margin Trend Over Time - (T)')
plt.xlabel('Date')
plt.ylabel('Net Profit Margin')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# Plotting the Return On Equity (ROE) and Return On Assets (ROA), alongside the dates, for Company 1 (Ford)
plt.figure(figsize=(10, 6))
plt.plot(fordMergeData['Date'], fordMergeData['ROE - Return On Equity'], label='ROE', marker='o', linestyle='-')
plt.plot(fordMergeData['Date'], fordMergeData['ROA - Return On Assets'], label='ROA', marker='o', linestyle='-')
plt.title('Profitability Ratios - (F)')
plt.xlabel('Date')
plt.ylabel('Ratio (%)')
plt.xticks(rotation=45)
plt.ylim(-100, 125) 
plt.legend()
plt.grid(True)
plt.show()

# Plotting the Return On Equity (ROE) and Return On Assets (ROA), alongside the dates, for Company 2 (Mercedes-Benz)
plt.figure(figsize=(10, 6))
plt.plot(benzMergeData['Date'], benzMergeData['ROE - Return On Equity'], label='ROE', marker='o', linestyle='-')
plt.plot(benzMergeData['Date'], benzMergeData['ROA - Return On Assets'], label='ROA', marker='o', linestyle='-')
plt.title('Profitability Ratios - (M)')
plt.xlabel('Date')
plt.ylabel('Ratio (%)')
plt.ylim(-10,10) 
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()


# Plotting the Return On Equity (ROE) and Return On Assets (ROA), alongside the dates, for Company 3 (Tesla)
plt.figure(figsize=(10, 6))
plt.plot(teslaMergeData['Date'], teslaMergeData['ROE - Return On Equity'], label='ROE', marker='o', linestyle='-')
plt.plot(teslaMergeData['Date'], teslaMergeData['ROA - Return On Assets'], label='ROA', marker='o', linestyle='-')
plt.title('Profitability Ratios - (T)')
plt.xlabel('Date')
plt.ylabel('Ratio (%)')
plt.ylim(-200, 420) 
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the expenses and revenue, alongside the dates, for Company 1 (Ford)
plt.figure(figsize=(10, 6))
plt.plot(fordMergeData['Date'], fordMergeData['Operating Expenses'], label='Operating Expenses', marker='o', linestyle='-')
plt.plot(fordMergeData['Date'], fordMergeData['Revenue'], label='Revenue', marker='o', linestyle='-')
plt.plot(fordMergeData['Date'], fordMergeData['Inventory Turnover Ratio'], label='Inventory Turnover Ratio', marker='o', linestyle='-')
plt.title('Profitability Drivers - (F)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Plotting the expenses and revenue, alongside the dates, for Company 2 (Mercedes-Benz)
plt.figure(figsize=(10, 6))
plt.plot(benzMergeData['Date'], benzMergeData['Operating Expenses'], label='Operating Expenses', marker='o', linestyle='-')
plt.plot(benzMergeData['Date'], benzMergeData['Revenue'], label='Revenue', marker='o', linestyle='-')
plt.plot(benzMergeData['Date'], benzMergeData['Inventory Turnover Ratio'], label='Inventory Turnover Ratio', marker='o', linestyle='-')
plt.title('Profitability Drivers - (M)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Plotting the expenses and revenue, alongside the dates, for Company 3 (Tesla)
plt.figure(figsize=(10, 6))
plt.plot(teslaMergeData['Date'], teslaMergeData['Operating Expenses'], label='Operating Expenses', marker='o', linestyle='-')
plt.plot(teslaMergeData['Date'], teslaMergeData['Revenue'], label='Revenue', marker='o', linestyle='-')
plt.plot(teslaMergeData['Date'], teslaMergeData['Inventory Turnover Ratio'], label='Inventory Turnover Ratio', marker='o', linestyle='-')
plt.title('Profitability Drivers - (T)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
