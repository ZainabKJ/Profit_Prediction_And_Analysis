```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import os
```

```python
# Read the orginal merge files from necessary folder
ford_merge = '../csv2/fMerge.csv'
benz_merge = '../csv2/mMerge.csv'
tesla_merge = '../csv2/tMerge.csv'

frame_f = pd.read_csv(ford_merge)
frame_b = pd.read_csv(benz_merge)
frame_t = pd.read_csv(tesla_merge)

# Implement scaler for normalisation
scaler = MinMaxScaler()

# Specifise the needed columns
column_value_normal = ['Current Ratio',
'Debt/Equity Ratio',
'Gross Margin',
'EBIT Margin',
'EBITDA Margin',
'Net Profit Margin',
'Inventory Turnover Ratio',
'Days Sales In Receivables',
'ROE - Return On Equity',
'ROA - Return On Assets',
'Book Value Per Share',
'Free Cash Flow Per Share',
'Cash On Hand',
'Receivables',
'Inventory',
'Pre-Paid Expenses',
'Other Current Assets',
'Total Current Assets',
'Property, Plant, And Equipment',
'Long-Term Investments',
'Goodwill And Intangible Assets',
'Other Long-Term Assets',
'Total Long-Term Assets',
'Total Assets',
'Total Current Liabilities',
'Long Term Debt',
'Other Non-Current Liabilities',
'Total Long Term Liabilities',
'Total Liabilities',
'Common Stock Net',
'Retained Earnings (Accumulated Deficit)',
'Comprehensive Income',
'Other Share Holders Equity',
'Share Holder Equity',
'Total Liabilities And Share Holders Equity',
'Revenue',
'Gross Profit',
'Research And Development Expenses',
'Operating Expenses',
'Operating Income',
'Income After Taxes',
'Net Income',
'EBITDA',
'EBIT',
'Total Non-Operating Income/Expense',
'Net Income/Loss',
'Cash Flow From Operating Activities',
'Cash Flow From Financial Activities',
'Cash Flow From Investing Activities',
'Net Cash Flow']  

# Normalize the chosen columns
normal_f = frame_f.copy() 
normal_b = frame_b.copy()
normal_t = frame_t.copy()

normal_f[column_value_normal] = scaler.fit_transform(normal_f[column_value_normal])
normal_b[column_value_normal] = scaler.fit_transform(normal_b[column_value_normal])
normal_t[column_value_normal] = scaler.fit_transform(normal_t[column_value_normal])

# Creation and implementation of folder for normalised files
final_fold_f = '../csv2/Normal/'
final_fold_b = '../csv2/Normal/'
final_fold_t = '../csv2/Normal/'

os.makedirs(final_fold_f, exist_ok=True)
os.makedirs(final_fold_b, exist_ok=True)
os.makedirs(final_fold_t, exist_ok=True)

# CSV file path 
csv_file_f = os.path.join(final_fold_f, 'fNormal.csv')  
csv_file_b = os.path.join(final_fold_b, 'bNormal.csv') 
csv_file_t = os.path.join(final_fold_t, 'tNormal.csv') 

# Normalised files saved
normal_f.to_csv(csv_file_f, index=False)
normal_b.to_csv(csv_file_b, index=False)
normal_t.to_csv(csv_file_t, index=False)
```
